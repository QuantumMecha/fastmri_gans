```python
!pip install s3fs
!pip install pydicom
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
plt.rcParams["figure.figsize"] = (10, 10)
import h5py
import s3fs
import pydicom
from pydicom.data import get_testdata_file
import gc  # Garbage collection
```

Connect to the S3 bucket with the brain fastMRI data.

```python
cols = ['Folder', 'File', 'SeriesDescription', 'Manufacturer', 'StudyDescription', 'Height', 'Width']

s3 = s3fs.S3FileSystem(anon=True)
dicom_folders = s3.ls('s3://cs7643-fastmri/data/fastMRI_brain_DICOM/')

dicom_df = pd.DataFrame(columns=cols)

num_folders = 0
num_files = 0
for folder in dicom_folders:
  dicom_files = s3.ls('s3://'+folder+'/')
  num_folders += 1
  for file in dicom_files:
    filename = 's3://'+file
    data = pydicom.dcmread(s3.open(filename, 'rb'))
    df = pd.DataFrame([[folder, file, data.SeriesDescription, data.Manufacturer, data.StudyDescription, data.pixel_array.shape[0], 
                        data.pixel_array.shape[1]]], columns=cols)
    dicom_df = dicom_df.append(df)
    num_files += 1
    if num_files % 100 == 0:
      print('Total of ', num_files, ' files in ', num_folders, ' folders')
      dicom_df.to_csv('/content/drive/My Drive/CS7643_Project/DICOM.csv', header=True, sep='|', index=False)
      del data
      del df
      gc.collect()

print('Total of ', num_files, ' files in ', num_folders, ' folders')
```
